```python
country_resp=df['Country'].value_counts()
```

```python
country_uses_python=country_grp['LanguageWorkedWith'].apply(lambda x:x.str.contains('Python').sum())
```

```python
country_uses_python
```

```python
python_df=pd.concat([country_resp,country_uses_python],axis='columns',sort=False)
```

```python
python_df
```

```python
python_df.rename(columns={'Country':'NumResp','LanguageWorkedWith':'NumKnowsPy'},inplace=True)
```

```python
python_df['PctKnowsPy']=(python_df['NumKnowsPy']/python_df['NumResp'])*100
```

```python
python_df.sort_values(by='PctKnowsPy',ascending=False, inplace=True)
```
